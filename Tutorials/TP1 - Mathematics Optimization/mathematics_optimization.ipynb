### Exercise 1: Analytical Determination of Critical Points

Determine _analytically_ the coordinates and nature of the critical points of the following functions:

- $f_1(x,y) = x^3 + y^3 -3x -12y$
- $f_2(x,y) = x^3 -12xy + 8y^3$
- $f_3(x,y) = {(x-1)}^2 + 10 {(x^2-y)}^2$

```python
# Import numpy and pyplot libraries
import numpy as np
import math
import pandas as pd 
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def f1(x, y):
    return x**3 + y**3 - 3*x - 12*y

def f2(x, y):
    return x**3 - 12*x*y + 8*y**3

def f3(x, y):
    return (x - 1)**2 + 10*(x**2 - y)**2
```

```python
# Define symbols
x, y = sp.symbols('x y', real=True)

def critical_points(f):
    """
    Compute critical points of a function f(x, y)
    Returns the solutions of grad f = 0 as a list of dicts.
    """

    # Compute gradient
    fx = sp.diff(f, x)
    fy = sp.diff(f, y)

    # Solve system fx = 0, fy = 0
    solutions = sp.solve([fx, fy], [x, y], dict=True)
    return solutions
```

```python
# Define the functions symbolically
f1_sym = x**3 + y**3 - 3*x - 12*y
f2_sym = x**3 - 12*x*y + 8*y**3
f3_sym = (x - 1)**2 + 10*(x**2 - y)**2

# Compute critical points
cp_f1 = critical_points(f1_sym)
cp_f2 = critical_points(f2_sym)
cp_f3 = critical_points(f3_sym)

print("Critical points of f1:", cp_f1)
print("Critical points of f2:", cp_f2)
print("Critical points of f3:", cp_f3)
```

#### Determination & Analysis of the Critical Points

**[1]** For function $f_1(x,y)=x^3+y^3-3x-12y$

**Solve:**
- $f_x = 3(x^2-1)=0$ $\implies$ $x=\pm 1$
- $f_y = 3(y^2-4)=0$ $\implies$ $y=\pm 2$

**Critical points:**
- $(x_1,y_1) = (1,2)$
- $(x_2,y_2) = (-1,-2)$
- $(x_3,y_3) = (1,-2)$
- $(x_4,y_4) = (-1,2)$

**Hessian matrix:**
- $H = \begin{bmatrix} 6x & 0 \\ 0 & 6y \end{bmatrix}$

**Conclusion:**
| Point     | $f_{xx}$ | $f_{yy}$ | Nature        |
|-----------|----------|----------|---------------|
| $(1,2)$   | >0       | >0       | Local minimum |
| $(1,-2)$  | >0       | <0       | Saddle point  |
| $(-1,2)$  | <0       | >0       | Saddle point  |
| $(-1,-2)$ | <0       | <0       | Local maximum |

**[2]** For function $f_2(x,y)=x^3-12xy+8y^3$

**Solve:**
- $f_x = 3x^2-12y = 0$ $\implies$ $y=\frac{1}{4}x^2$
- $f_y = -12x+24y^2 = 0$

**Substitute:**
- $f_y = -12x+24y^2 = 0$ $\implies$ $x(6x^3-12) = 0$
- $x=0$ or $x=2$ $\implies$ $y=0$ or $y=1$

**Critical points:**
- $(x_1,y_1) = (0,0)$
- $(x_2,y_2) = (2,1)$

**Hessian matrix:**
- $H(f_2) = \begin{bmatrix} 6x & -12 \\ -12 & 48y \end{bmatrix}$

**Conclusion:**
| Point     | $f_{xx}$ | Hessian determinant | Nature        |
|-----------|----------|---------------------|---------------|
| $(0,0)$   | <0       | â€”                   | Saddle point  |
| $(2,1)$   | >0       | >0                  | Local minimum |

**[3]** For function $f_3(x,y) = {(x-1)}^2 + 10 {(x^2-y)}^2$

**Solve:**
- $f_y = -20(x^2-y)=0$ $\implies$ $y=x^2$
- $f_x = 2(x-1)+40x(x^2-y)=0$

**Substitute:**
- $f_x = 2(x-1)+40x(x^2-y)=0$ $\implies$ $2(x-1) = 0$
- $x=1$ $\implies$ $y=1$

**Critical Point:**
- $(x,y) = (1,1)$

**Hessian:**
- $H = \begin{bmatrix} 82 & -40 \\ -40 & 20 \end{bmatrix}$

where $det(H) = 40 > 0$ and $f_{xx} = 82 > 0$

**Conclusion:**
- $(x,y) = (1,1)$ is a local minimum
- Because $f_3$  is a convex function, it is also the global minimum.

### Exercise 3: Plotting Contour Lines (Isovalues) of a Function of Two Variables

**3.1** The instruction block below allows you to plot the contour lines (isovalues) of the function $f_1(x,y)$ for $x \in [-4,4]$ and $y \in [-4,4]$. 

Vary the different plotting parameters (intervals of x and y, number of discretization points, number of contour lines) and analyze their influence. In particular, what happens if the domain discretization is insufficient?

```python
# Different plotting parameter sets with more variation
params = [
    {'xmin': -4, 'xmax': 4, 'nx': 11, 'ymin': -4, 'ymax': 4, 'ny': 11, 'nIso': 10},
    {'xmin': -4, 'xmax': 4, 'nx': 81, 'ymin': -4, 'ymax': 4, 'ny': 81, 'nIso': 50},
    {'xmin': -10, 'xmax': 10, 'nx': 41, 'ymin': -10, 'ymax': 10, 'ny': 41, 'nIso': 21},
    {'xmin': -2, 'xmax': 2, 'nx': 41, 'ymin': -2, 'ymax': 2, 'ny': 41, 'nIso': 15},
    {'xmin': -8, 'xmax': 6, 'nx': 41, 'ymin': -4, 'ymax': 10, 'ny': 41, 'nIso': 20},
    {'xmin': 0, 'xmax': 4, 'nx': 41, 'ymin': -4, 'ymax': 0, 'ny': 41, 'nIso': 25}
]

# Create subplot grid
fig, axes = plt.subplots(3, 2, figsize=(14, 16))
axes = axes.flatten()

for i, p in enumerate(params):
    x1d = np.linspace(p['xmin'], p['xmax'], p['nx'])
    y1d = np.linspace(p['ymin'], p['ymax'], p['ny'])
    x2d, y2d = np.meshgrid(x1d, y1d)
    
    cs = axes[i].contour(x2d, y2d, f1(x2d, y2d), p['nIso'])
    axes[i].set_title(f"nx={p['nx']}, ny={p['ny']}, nIso={p['nIso']}\n"
                      f"x in [{p['xmin']},{p['xmax']}], y in [{p['ymin']},{p['ymax']}]")
    axes[i].set_xlabel('x values')
    axes[i].set_ylabel('y values')
    axes[i].grid()
    axes[i].axis('square')

plt.tight_layout()
plt.show()
```

Analysis of the parameters:

- **Discretization points (`nx`, `ny`):** Increasing the number of points makes the contours smoother and more precise, especially for regions where the function changes rapidly.

- **Contour levels (`nIso`):** More contour lines provide a finer representation of the functionâ€™s variations, while fewer lines highlight only the main trends.

- **Domain (`xmin`, `xmax`, `ymin`, `ymax`):** Zooming in or out changes which features are visible. Large domains show the overall shape, small domains focus on local structures.

If the domain discretization is insufficient (meaning `nx` and `ny` are too small), the contour plot will not accurately capture the variations of the function. Specially:

- **Contours appear jagged or blocky** because there are too few points to interpolate smooth curves.
- **Small features may be missed** entirely, such as local maxima, minima, or saddle points.
- **Level sets can be misrepresented**, potentially giving a misleading impression of the functionâ€™s topology.
- **Numerical artifacts may appear**, especially in regions where the function changes rapidly.

In short, coarse discretization reduces the fidelity of the visualization, making it harder to correctly interpret the functionâ€™s behavior. For accurate contour plots, the grid needs to be fine enough to resolve the functionâ€™s important features.

**3.3** The instruction block below provides the template for the function _df1_, which calculates and returns the gradient of $f_1$. Complete this instruction block so that _df1_ does what is expected. Then use this function to verify that the gradient of $f_1$ is indeed zero at the critical points.

```python
def df1(x, y):
    # Partial derivative with respect to x
    df_dx = 3*x**2 - 3
    # Partial derivative with respect to y
    df_dy = 3*y**2 - 12
    return [df_dx, df_dy]
```

```python
# Critical points
critical_points_f1 = [(1, 2), (-1, -2), (-1, 2), (1, -2)]

# Check gradient at each point
for point in critical_points_f1:
    grad = df1(point[0], point[1])
    print(f"Gradient at {point}: {grad}")
```

**3.4** The instruction block below provides the template for the function _d2f1_, which calculates and returns the Hessian matrix of $f_1$. Complete this instruction block so that _d2f1_ does what is expected. Then use this function to verify the nature of the critical points of $f_1$.

```python
def d2f1(x, y):
    # Second-order partial derivatives
    d2f_dxdx = 6*x
    d2f_dxdy = 0
    d2f_dydx = 0
    d2f_dydy = 6*y
    
    return [[d2f_dxdx, d2f_dxdy],
            [d2f_dydx, d2f_dydy]]
```

```python
for (x, y) in critical_points_f1:

    # Hessian matrix at (x, y)
    H = d2f1(x, y)
    eigenvalues = np.linalg.eigvals(H)

    print(f"\n=== Critical point (x={x}, y={y}) ===")
    print("Hessian matrix:")
    print(H)

    print("Eigenvalues:")
    print(f"  Î»â‚ = {eigenvalues[0]:.4f}")
    print(f"  Î»â‚‚ = {eigenvalues[1]:.4f}")

    # Nature of critical point
    if np.all(eigenvalues > 0):
        nature = "Local minimum"
    elif np.all(eigenvalues < 0):
        nature = "Local maximum"
    else:
        nature = "Saddle point"

    print(f"Nature of the point: {nature}")
```
